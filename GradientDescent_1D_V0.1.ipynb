```python
import numpy as np
import sympy as sym
import random
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
def expr(x):
  return np.cos(2*np.pi*x) + x**2
def deriv(x):
  return 2*x - 2*np.pi*np.sin(np.pi*2*x)
```

```python
x = np.linspace(-2,2,2000)
localmin = np.random.choice(x,1)
print(localmin)

learning_rate = 0.01
learning_epochs = 1000

for i in range(learning_epochs):
  localmin = localmin - learning_rate*deriv(localmin)

plt.plot(x, expr(x), 'r--')
plt.plot(x, deriv(x), 'b-')
plt.plot(localmin, expr(localmin), 'ro')
plt.xlim(x[[0,-1]])
plt.xlabel("iteration")
plt.ylabel("localmin")
plt.title("Gradiant descent")
localmin
```
